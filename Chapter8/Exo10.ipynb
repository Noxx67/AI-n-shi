```python
from sklearn.datasets import fetch_openml

X_mnist, y_mnist = fetch_openml("mnist_784", as_frame=False, return_X_y=True)
```

```python
from sklearn.manifold import TSNE

random_bs = TSNE(2, random_state=67, n_jobs=-1)

X2D_mnist = random_bs.fit_transform(X_mnist)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

plt.scatter(X2D_mnist[:,0], X2D_mnist[:,1], alpha=0.2, cmap='jet', c=y_mnist.astype(int))
plt.show()

#das so cool
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca_reduced = pca.fit_transform(X_mnist)
plt.scatter(X_pca_reduced[:,0], X_pca_reduced[:,1], alpha=0.2, cmap='jet', c=y_mnist.astype(int))
plt.show() #hmar
```

```python
X_sample = X_mnist[:3000]
y_sample = y_mnist[:3000]
```

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, random_state=42)
%time X_lle_reduced = lle.fit_transform(X_sample)
plt.scatter(X_lle_reduced[:,0], X_lle_reduced[:,1], alpha=0.2, cmap='jet', c=y_sample.astype(int))
plt.show()
```
