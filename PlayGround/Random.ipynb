```python
from sklearn.datasets import fetch_openml

X_mnist, y_mnist = fetch_openml("mnist_784", as_frame=False, return_X_y=True)
```

```python
X_train, y_train = X_mnist[:60000], y_mnist[:60000]
X_test, y_test = X_mnist[60000:], y_mnist[60000:]
```

```python
y_train_5 = (y_train == '5')  # True for all 5s, False for all other digits
y_test_5 = (y_test == '5')
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import recall_score

hmar = DummyClassifier()

hmar.fit(X_train, y_train_5)
recall_score(y_test_5, hmar.predict(X_test))
```
