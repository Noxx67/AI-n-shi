```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine(as_frame=True)

X = wine.data
y = wine.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=67)
```

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier #book said to use this since SVC uses OvO automatically
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# model = SVC(random_state=67)
model = OneVsRestClassifier(SVC(random_state=67))
acc = cross_val_score(model, X_scaled, y_train, cv=3, scoring="accuracy")
acc.mean() #98.58 acc ???
model.fit(X_scaled, y_train)
predictions = model.predict(scaler.transform(X_test))
print("cross validation score: ", acc.mean())
print("validation score: ", accuracy_score(y_test, predictions)) #100 acc ????????????????????????????????
```

```python
#doing gridsearch because fuck it
from sklearn.model_selection import GridSearchCV

param_grid = {
    'estimator__C': [0.1, 1, 10],       
    'estimator__kernel': ['rbf'],  
    'estimator__gamma': ['scale', 'auto']     
}

grid = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)

grid.fit(X_scaled, y_train)

print("Best params:", grid.best_params_) # C=0.1, rbf, scale
print("Best cross-val score:", grid.best_score_) # 0.99?!?!?!?!??

predictions = grid.predict(scaler.transform(X_test))
print("best score: ", accuracy_score(y_test, predictions)) #1 because yes
```

```python
the_goat_ig = grid.best_estimator_

final_predictions_because_im_suspicious_of_the_results = the_goat_ig.predict(scaler.transform(X))
print("final acc:", accuracy_score(y,final_predictions_because_im_suspicious_of_the_results))
#0.994 acc which is crazy
```
